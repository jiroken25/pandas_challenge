```python
import pandas as pd
import os

pd.options.display.precision = 2
```

```python
csvpath = os.path.join('resources', 'purchase_data.csv')


df = pd.read_csv(csvpath)

```

```python
# Confirm the item number
df.info()
df.head()
```

```python
# Player Count
# Total Number of Players
totalplayers = len(df["SN"].unique())
totalplayers_df = pd.DataFrame({"Total Players":[totalplayers]})
totalplayers_df
```

```python
# Purchasing Analysis (Total)

# Number of Unique Items
itemslist = df["Item ID"].unique()

# Average Purchase Price
price_mean = df["Price"].mean()

# Total Number of Purchases
itemIDlist = df["Item ID"]

# Total Revenue
totalrevenue = df["Price"].sum()


purchase_analysis_table = pd.DataFrame({"Number of Unique Items":'{0:d}'.format(len(itemslist)),"Average Purchase Price":price_mean,"Total Number of Purchaes":['{0:d}'.format(len(itemIDlist))], "Total Revenue":totalrevenue})
purchase_analysis_table
```

```python
# Gender Demographics
# Percentage and Count of Male Players
player_count = len(df["SN"].unique())
maledf = df[df["Gender"] == "Male"]
male_count = len(maledf["SN"].unique())     


# Percentage and Count of Female Players
femaledf = df[df["Gender"] == "Female"]
female_count = len(femaledf["SN"].unique())

# Percentage and Count of Other / Non-Disclosed
otherdf = df[df["Gender"] == "Other / Non-Disclosed"]
other_count = len(otherdf["SN"].unique())
# Purchasing Analysis (Gender)

purchase_gender_table = pd.DataFrame({"Gender":["Female","Male","Other / Non-Disclosed"],
                                     'Count of players':[female_count, male_count, other_count],
                                    'Percentage%':['{0:.2f}%'.format(100*female_count/player_count), '{0:.2f}%'.format(100*male_count/player_count), '{0:.2f}%'.format(100*other_count/player_count)]})
purchase_gender_table
 
```

```python
# The below each broken by gender

# Purchase Count
gender_pcount = df[["Gender","Price"]].groupby("Gender").count().reset_index()["Price"]


# Average Purchase Price
gender_pmean = df[["Gender","Price"]].groupby("Gender").mean().reset_index()["Price"]
# Total Purchase Value
# Average Purchase Total per Person by Gender
gender_df = df[["Gender","Price"]].groupby("Gender").sum().reset_index()
gender_df["Unique user count"] = [female_count,male_count,other_count]
gender_df["Average Purchase Total per Person by Gender"] = gender_df["Price"]/ gender_df["Unique user count"]

gender_df["Purchase Count"] = gender_pcount
gender_df["Average Purchase Price"] = gender_pmean

gender_df = gender_df.rename(columns={"Price":"Total Purchase Value"})

gender_df = gender_df[["Gender","Purchase Count","Average Purchase Price","Total Purchase Value","Unique user count","Average Purchase Total per Person by Gender"]]

gender_df


```

```python
age_list = [0,9,14,19,24,29,34,39,44,49]
age_label = [" - 9", "10-14","15-19","20-24","25-29","30-34","35-39","40-44","45 -"]

df["Age Range"] = pd.cut(df["Age"],bins=age_list, labels=age_label)

dic_age = {"Age Range":age_label,"Player Count":[],"Percentage%":[]}

for age in age_label:
    dic_age["Player Count"].append(len(df[df["Age Range"] == age]["SN"].unique()))
    dic_age["Percentage%"].append('{0:.2f}%'.format(len(df[df["Age Range"] == age]["SN"].unique())/player_count*100))

age_percent_df = pd.DataFrame.from_dict(dic_age)
age_percent_df

```

```python
# Age Demographics
# Get the statistic info using describbe
# df["Age"].describe()
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
# 7-45 should be age_label = ["< 9", "10-14","15-19","20-24","25-29","30-34","35-39","40-44"]


dic_store = {"Age Range":age_label,"Purchase Count":[], "Average Purchase Price":[],"Total Purchase Price":[],"Unique user count":[],"Average Purchase Total per Person by Age Group":[]}


# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Age Group
# Use for loop to repeat all age dataframe

for age in age_label:
    dic_store["Purchase Count"].append(df[df["Age Range"] == age]["Purchase ID"].count())
    dic_store["Average Purchase Price"].append(df[df["Age Range"] == age]["Price"].mean())
    dic_store["Total Purchase Price"].append(df[df["Age Range"] == age]["Price"].sum())
    dic_store["Unique user count"].append(len(df[df["Age Range"] == age]["SN"].unique()))
    dic_store["Average Purchase Total per Person by Age Group"].append(df[df["Age Range"] == age]["Price"].sum()/len(df[df["Age Range"] == age]["SN"].unique()))


    
age_df = pd.DataFrame.from_dict(dic_store)
age_df
```

```python
# Top Spenders

# Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

topfivename = df.groupby("SN").sum()["Price"].sort_values(ascending = False).head(5).reset_index()["SN"]
# SN
# Purchase Count
# Average Purchase Price
# Total Purchase Value

store2_dic = {"SN":[],"Purchase Count":[],"Average Purchase Price":[],"Total Purchase Price":[]}
for name in topfivename:
    store2_dic["SN"].append(name)
    store2_dic["Purchase Count"].append(df[df["SN"] == name]["Purchase ID"].count())
    store2_dic["Average Purchase Price"].append(df[df["SN"] == name]["Price"].mean())
    store2_dic["Total Purchase Price"].append(df[df["SN"] == name]["Price"].sum())
        
spender_df = pd.DataFrame.from_dict(store2_dic)
spender_df








```

```python
# Most Popular Items

toppopitemID = df.groupby("Item ID").count().reset_index()[["Item ID","Price"]].sort_values("Price",ascending = False).head(5)["Item ID"]

# Identify the 5 most popular items by purchase count, then list (in a table):
store3_dic = {"Item ID":[],"Item Name":[],"Purchase Count":[],"Item Price":[],"Total Purchase Value":[]}
for ID in toppopitemID:
    store3_dic["Item ID"].append(ID)
    store3_dic["Item Name"].append(df[df["Item ID"] == ID]["Item Name"].unique()[0])
    store3_dic["Purchase Count"].append(df[df["Item ID"] == ID]["Purchase ID"].count())
    store3_dic["Item Price"].append(df[df["Item ID"] == ID]["Price"].mean())
    store3_dic["Total Purchase Value"].append(df[df["Item ID"] == ID]["Price"].sum())

popular_df = pd.DataFrame.from_dict(store3_dic)
popular_df
# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value
```

```python
# Most Profitable Items
topprofitemID = df.groupby("Item ID").sum()["Price"].sort_values(ascending = False).reset_index().head(5)["Item ID"]

# Identify the 5 most profitable items by total purchase value, then list (in a table):
store4_dic = {"Item ID":[],"Item Name":[],"Purchase Count":[],"Item Price":[],"Total Pruchase Value":[]}
for ID in topprofitemID:
    store4_dic["Item ID"].append(ID)
    store4_dic["Item Name"].append(df[df["Item ID"] == ID]["Item Name"].unique()[0])
    store4_dic["Purchase Count"].append(df[df["Item ID"] == ID]["Purchase ID"].count())
    store4_dic["Item Price"].append(df[df["Item ID"] == ID]["Price"].mean())
    store4_dic["Total Pruchase Value"].append(df[df["Item ID"] == ID]["Price"].sum())

profit_df = pd.DataFrame.from_dict(store4_dic)
profit_df

# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value
```
